```python
Portfolio2 = pd.concat([df_AMC, df_ANZ, df_BHP, df_BXB, df_CBA, 
                        df_CSL, df_GMG, df_IAG, df_MQG, df_NAB,
                        df_NCM, df_RIO, df_SCG, df_SUN, df_TLS, df_TCL, 
                        df_WES, df_WBC, df_WPL, df_WOW], axis="columns", join="inner")

Portfolio2.columns = ['AMC', 'ANZ' , 'BHP', 'BXB', 'CBA' , 'CSL', 'GMG', 'IAG', 'MQG', 'NAB', 'NAM', 'RIO' , 'SCG' , 'SUN', 'TLS', 'TCL', 'WES', 'WBC', 'WPL', 'WOW']
Portfolio2
```

```python
# Optimize the portfolio 
# Caluculate the expected anualized returns and sample covariance

mu = expected_returns.mean_historical_return(Portfolio2)
S = risk_models.sample_cov(Portfolio2)
mu
```

```python
#Optimized for the miximal Sharpe ratio 
ef = EfficientFrontier(mu, S)  #create the Efficient Frontier Object
weights = ef.max_sharpe()

cleaned_weights = ef.clean_weights()
ef.save_weights_to_file("weights.txt")  # saves to file
print(cleaned_weights)
```

```python
ef.portfolio_performance(verbose = True)
```

```python
# pip install pulp
# pip install cvxopt (FOR solver GLPK_MI)

# Get the discrete allocation of each share per share per stock 
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices

initial_investment = 10000
latest_prices = get_latest_prices(Portfolio2)
weights = cleaned_weights 

da = DiscreteAllocation(weights, latest_prices, total_portfolio_value = initial_investment)
allocation, leftover = da.lp_portfolio()
print('Discrete allocation', allocation)
print('Funds Remaining: $' , leftover)
```
